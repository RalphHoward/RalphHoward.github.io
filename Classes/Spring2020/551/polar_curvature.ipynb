## This notebook both plots an equation $r= f(\theta)$ in polar coordinates and computes its curvature.

Start by importing the required libraries:

```python
from sympy import *  #python symbolic math package.
import numpy as np   #numerical mathmatics library.
import matplotlib.pyplot as plt    #plotting library.
init_printing()            #to get good looking output.
```

To make things a bit easier input with will use $t=\theta$ 
for the variable.  The code below declares $t$ to be a
symbol, gives the domain, $[a,b]$, of the function $f$
and defines $f(t)$.  The definition of $f$ should be
edited as needed.  Then, as we will need these values
to plot the graph, define give the formulas for $x$ and $y$.

```python
t = symbols('t')
a = 0
b = 2*pi
f = 1 - 2 * cos(t) #default.  Edit as needed. 
x = f*cos(t)
y = f*sin(t)
```

Divide the domain into pieces of length $h$ to be used in plotting
and compute the $x$ and $y$ values at these subdivsions of the domain
and store in them in lists.

```python
h = .01   #edit if need be.
t_dom = np.arange(a,b,h)
x_vals = [x.subs(t,s) for s in t_dom]
y_vals = [y.subs(t,s) for s in t_dom]
```

Plot the curve.

```python
plt.plot(x_vals,y_vals)
plt.axis('equal')
```

Compute the speed $v$.

```python
df = diff(f,t)
v = (f**2 + df**2)**(1/2)
v.simplify()  # simplfy v if possible.
```

Compute the curvature.

```python
ddf = diff(df,t)
numerator = simplify(f*f + 2*df*df - f*ddf)
numerator.simplify()  # simplity some more
kappa = numerator/v**3
kappa.simplify()
```

Plot the curvature.

```python
kappa_vals = [kappa.subs(t,s) for s in t_dom]
plt.plot(t_dom,kappa_vals)
```
